## 4.0 Volume of the Universe covered by the survey

The cosmological volume $V_C$ for a given redshift is:

$$V_C = 4/3 \pi d_p^3,$$

where $d_p$ is the comoving distance of the galaxy. 

The volume contained between redshifts $z_1$ and $z_2$ by an $X$-square-degree survey is given by:

$$\mathrm{V_S = \frac{X}{4\pi (180/\pi)^2} [V_C(z_2) - V_C(z_1)]}$$

```python
dp = Planck18.comoving_distance(0.1) # find the co-moving distance at z = 0.1 (in Mpc)
Vc = (4/3) * np.pi * dp**3 # in Mpc^3
Vs = (10 / (4*np.pi * (180/np.pi)**2 )) * Vc
galaxy_sample['volume'] = [(10 / (4*np.pi * (180/np.pi)**2 )) * (4/3) * np.pi * Planck18.comoving_distance(z)**3 for z in galaxy_sample['Redshift']]
print(f"Survey volume, Vs = {Vs:.0f}")
```

## 5.0 Luminosity distance

```python
galaxy_sample['luminosity distance'] = [Planck18.luminosity_distance(z) for z in galaxy_sample['Redshift']]
```

## 6.0 Absolute magnitude

The absolute magnitude is given by:

$$M = m - 5\log_{10}\left(\frac{d}{10 (pc)}\right),$$
where $m$ is the apparent magnitude and $d$ is the luminosity distance in parsecs.

```python
def absolute_magnitude(m, m_err, d): # apparent mag, apparent dmag, luminosity distance
    """
    calculate the absolute magnitude of a source given its apparent magnitude and the luminosity distance
    """
    # check the units of d (change to pc if not already)
    if d.unit != u.pc:
        d = d.to("pc")
    # calculate absolute magnitude
    M = m - 5 * np.log10(d.value/10)
    # calculate error
    M_max = (m+m_err) - 5 * np.log10(d.value/10)
    M_err = M_max - M
    return M, M_err
```

```python
absmag = []
d_absmag = []
for i in galaxy_sample.index:
    m, dm = absolute_magnitude(galaxy_sample['K-band mag'][i],
                              galaxy_sample['K-band dmag'][i],
                              galaxy_sample['luminosity distance'][i]
                              )
    absmag.append(m)
    d_absmag.append(dm)
    
galaxy_sample['abs mag'] = absmag
galaxy_sample['abs dmag'] = d_absmag

plotsize()
plt.errorbar(galaxy_sample['Redshift'], galaxy_sample['abs mag'], galaxy_sample['abs dmag'], c='#0060aa', fmt='.', capsize=2, alpha=0.6)
plt.xlabel("Redshift"), plt.ylabel("Absolute magnitude")
plt.gca().invert_yaxis()
```

## 7.0 Limiting absolute magnitude

```python
# the limiting apparent magnitude is mag_lim
# at z = 0.1, the limiting apparent magnitude comes into play...
absmag_lim = absolute_magnitude(mag_lim, 0, Planck18.luminosity_distance(0.1))[0]
print(f"Limiting absolute magnitude = {absmag_lim:.1f}")
```

## 8.0 Galaxy luminosity function

We need to have the number density of galaxies per Mpc$^3$ per magnitude. To do this, we should put the number density per Mpc$^3$ into luminosity bins.

If we first place the number of galaxies into luminosity bins, then we will have the number density per magnitude. Then if we divide this number by the vurvey volume Vs, we have the number density per Mpc$^3$ per magnitude.

```python
# find the galaxy luminosities
galaxy_sample['luminosity'] = [10**((4.85-M)/2.5) for M in galaxy_sample['abs mag']]

# number of galaxies per absolute magnitude
galnum, galbins = np.histogram(galaxy_sample['abs mag'], bins=36)
mags = np.linspace(round(min(galaxy_sample['abs mag'])), round(max(galaxy_sample['abs mag'])), len(galnum))

plt.scatter(mags, galnum)
plt.xlabel("Absolute magnitude"), plt.ylabel("Number of galaxies"), plt.title("Number of galaxies per absolute magnitude")
```

```python
# number of galaxies per absolute magnitude per cubic comoving distance
galnumlum = galnum/Vs

plt.scatter(mags, galnumlum)
plt.xlabel("Absolute magnitude"), plt.ylabel(r"N$_{\mathrm{gals}}$ mag$^{-1}$ Mpc$^{-3}$"), plt.title("Number of galaxies per magnitude per cubic comoving distance")
```

```python
# log and solar luminosity scale
plt.scatter(np.log10( 10**((4.85-mags)/2.5 )), np.log10(galnumlum.value))
plt.xlabel(r"log$_{10}$( L / L$_\odot$ )"), plt.ylabel(r"log$_{10}$( N$_{\mathrm{gals}}$ ) mag$^{-1}$ Mpc$^{-3}$")
plt.title("Galaxy luminosity function?")
```

## 9. Schechter function

The Schechter function is given by:

$$n(M) dM = 0.4 \ln(10) \phi^* \left[ 10^{ 0.4 (M^* - M) } \right]^{\alpha+1} e^{ -10^{0.4(M^* - M)} } dM,$$

where $M^*$, $\phi^*$ and $\alpha$ should be found from the best fit to the luminosity function. 

```python
def schechter(magnitude, M, phi, alpha):
    return 0.4 * np.log(10) * phi * ( 10**(0.4*(M-magnitude))**(alpha+1) * np.exp( -10**(0.4*(M-magnitude)) ) )
```
